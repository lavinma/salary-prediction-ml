```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Results Table
results = pd.DataFrame({
    "Model": [
        "Dummy Regressor",
        "Linear Regression",
        "Random Forest"
    ],
    "MAE": [
        44175.28,
        32594.01,
        28043.28
    ],
    "RMSE": [
        61978.15,
        50656.20,
        47188.25
    ],
    "R²": [
        0.000,
        0.332,
        0.420
    ]
})

results
```

## Model Performance Comparison

The table below summarizes the performance of each regression model evaluated during this project.

Lower MAE and RMSE indicate better predictive accuracy, while higher R² values indicate that a model explains a greater proportion of the variation in annual salary.

```python
# Model comparison visualization
plt.figure(figsize=(8,5))

bars = plt.bar(
    results["Model"],
    results["R²"]
)

plt.title("Model Comparison (R² Score)", fontsize=16)
plt.ylabel("R²")

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x()+bar.get_width()/2,
        height+0.01,
        f"{height:.3f}",
        ha="center"
    )

plt.ylim(0,0.5)

plt.show()
```

### Interpretation

Model performance improves steadily as more sophisticated algorithms are introduced.

The Dummy Regressor provides a baseline with essentially no predictive power. Linear Regression captures meaningful relationships between job characteristics and salary, explaining approximately one-third of the observed variation.

The Random Forest Regressor achieves the highest R² score and the lowest prediction errors, indicating that it is better able to model the nonlinear relationships present within the dataset.

These results demonstrate that machine learning models can leverage structured job posting information to produce meaningful salary predictions.

## Strengths of the Approach

Several aspects of this project contribute to the effectiveness of the modeling pipeline.

### Strengths

- Built from over 35,000 cleaned LinkedIn job postings with annual salary information.
- Followed a complete end-to-end machine learning workflow from raw data acquisition through model evaluation.
- Used reproducible preprocessing pipelines to avoid data leakage.
- Compared multiple regression algorithms rather than relying on a single model.
- Demonstrated that structured job posting characteristics contain meaningful predictive information.

## Limitations

Although the Random Forest model performed well, several limitations remain.

### Limitations

- Salary depends on many factors that are not available in the dataset, such as education, years of experience, interview performance, and negotiation.
- Job descriptions were not incorporated into the models.
- One-hot encoding creates thousands of sparse categorical features.
- Some salary estimates may contain reporting errors despite the cleaning process.
- The dataset represents LinkedIn postings and may not generalize perfectly to the broader labor market.

## Future Improvements

Future versions of this project could further improve predictive performance by incorporating additional data and more advanced modeling techniques.

Potential improvements include:

- Natural language processing of job descriptions using TF-IDF or transformer embeddings.
- Target encoding or frequency encoding for high-cardinality categorical variables.
- Gradient Boosting, XGBoost, or LightGBM models.
- Hyperparameter optimization using GridSearchCV or RandomizedSearchCV.
- Model deployment as an interactive salary prediction web application.
