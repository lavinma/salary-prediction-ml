## Feature Selection

Not every column is useful for predicting salary.

This project focuses on variables that describe the characteristics of each job posting while excluding identifiers, URLs, and variables that would introduce data leakage.

The target variable for prediction is:

**normalized_salary**

```python
# I will now create the modeling dataset

jobs = postings.copy()

jobs = jobs.dropna(subset=["normalized_salary"])

jobs.shape


model_columns = [
    "title",
    "location",
    "formatted_work_type",
    "formatted_experience_level",
    "company_name",
    "views",
    "applies",
    "normalized_salary"
]

ml_df = jobs[model_columns].copy()

ml_df.head()
```

```python
ml_df.isnull().sum().sort_values(ascending=False)
```

## Standardize Text Features

Before handling missing values, I standardize the categorical text features by removing any leading or trailing whitespace.

This helps ensure that categories differing only by extra spaces (for example, `"Data Engineer"` versus `" Data Engineer"`) are treated as the same value, reducing duplicate categories and improving the quality of the engineered features.

```python
text_columns = [
    "title",
    "location",
    "formatted_work_type",
    "formatted_experience_level",
    "company_name"
]

for col in text_columns:
    ml_df[col] = ml_df[col].str.strip()

ml_df.head()
```

## Data Cleaning

Prepare the selected features by handling missing values.

The preprocessing strategy includes:

- Replacing missing categorical values with `"Unknown"`
- Filling missing numerical values using the median
- Removing features with excessive missingness

```python
ml_df = ml_df.drop(columns=["applies"])
ml_df["company_name"] = ml_df["company_name"].fillna("Unknown")
ml_df["formatted_experience_level"] = (
    ml_df["formatted_experience_level"]
    .fillna("Unknown")
)
ml_df["views"] = ml_df["views"].fillna(
    ml_df["views"].median()
)
ml_df.isnull().sum()
```

## Remove Unrealistic Salary Values

During exploratory inspection of the salary data, I identified a small number of job postings with salary values that were clearly inconsistent with the listed positions and compensation information.

Examples included:

- Annual salaries exceeding **$100 million** for standard professional roles.
- Annual salaries below **$20,000**, including values such as **$1**, **$13**, and **$212.50** for full-time positions that were labeled as yearly compensation.

Further investigation of the original dataset showed that some of these records contained inconsistent or incorrectly normalized salary information. Because the goal of this project is to predict **annual salaries**, I removed observations with implausible salary values to improve data quality and reduce the influence of erroneous records on downstream analysis and machine learning models.

For this project, I retain annual salaries between **$20,000** and **$1,000,000**.

```python
MIN_REASONABLE_SALARY = 20_000
MAX_REASONABLE_SALARY = 1_000_000

rows_before = len(ml_df)

ml_df = ml_df[
    (ml_df["normalized_salary"] >= MIN_REASONABLE_SALARY) &
    (ml_df["normalized_salary"] <= MAX_REASONABLE_SALARY)
].copy()

rows_after = len(ml_df)

print(f"Rows before filtering: {rows_before:,}")
print(f"Rows after filtering: {rows_after:,}")
print(f"Rows removed: {rows_before - rows_after:,}")
```
